```python
import pandas as pd

df = pd.read_csv("ecommerce_behavior_data.csv")

df.head()
```

```python
df.info()
```

```python
funnel = df.groupby('event_type')['user_id'].nunique()
print(funnel)
```

```python
views = df[df['event_type']=='view']['user_id'].nunique()
purchases = df[df['event_type']=='purchase']['user_id'].nunique()

conversion_rate = (purchases / views) * 100
print("Conversion Rate:", round(conversion_rate,2), "%")
```

```python
df['event_time'] = pd.to_datetime(df['event_time'])
df['date'] = df['event_time'].dt.date
```

```python
daily_users = df.groupby('date')['user_id'].nunique()
daily_users.head()
```

```python
dates = sorted(df['date'].unique())

day1_users = set(df[df['date'] == dates[0]]['user_id'])
day2_users = set(df[df['date'] == dates[1]]['user_id'])

retention_rate = len(day1_users & day2_users) / len(day1_users) * 100

print("Day 1 to Day 2 Retention:", round(retention_rate,2), "%")
```
